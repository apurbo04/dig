```python
import os
import numpy as np
import librosa
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Dataset path
dataset_path = r"E:\Academic\32\ICE-3207 Digital Image and Speech Processing\ICE-3208 (Digital Image and Speech Processing Sessional)\speaker_dataset"

# Automatically detect speaker folders
speakers = os.listdir(dataset_path)

X = []
y = []

# Extract embedding from audio file
def extract_embedding(file_path):
    signal, sr = librosa.load(file_path, sr=None)

    mfcc = librosa.feature.mfcc(y=signal, sr=sr, n_mfcc=13)
    delta = librosa.feature.delta(mfcc)

    mfcc_mean = np.mean(mfcc, axis=1)
    delta_mean = np.mean(delta, axis=1)

    embedding = np.concatenate([mfcc_mean, delta_mean])

    return embedding


# Load dataset
for speaker in speakers:
    folder = os.path.join(dataset_path, speaker)

    if not os.path.isdir(folder):
        continue

    for file in os.listdir(folder):
        if file.endswith(".wav"):
            file_path = os.path.join(folder, file)

            embedding = extract_embedding(file_path)

            X.append(embedding)
            y.append(speaker)

X = np.array(X)
y = np.array(y)

# Split dataset into train/test
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
```

```python
# Train KNN classifier
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(X_train, y_train)

# Predict
y_pred = clf.predict(X_test)

# Evaluation
accuracy = accuracy_score(y_test, y_pred)

print("Speaker Identification Accuracy:", round(accuracy * 100, 2), "%")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
