```python
#lab-01
# Write a program to draw a raditation pattern of dipole Antenna 
# Dipole Antenna-2D 
import numpy as np
import matplotlib.pyplot as plt

# Angle values (0 to 180 degree) 
theta = np.linspace(0, np.pi, 360)

# Radiation pattern
R = np.abs(np.sin(theta))

# Polar plot
plt.figure()
ax = plt.subplot(111, projection='polar')
ax.plot(theta, R)

ax.set_title("2D Radiation Pattern of Dipole Antenna", va='top')
plt.show()
```

```python
# Dipole Antenna-3D 
import numpy as np
import matplotlib.pyplot as plt

# Angular values
theta = np.linspace(0, np.pi, 100)
phi = np.linspace(0, 2*np.pi, 100)
theta, phi = np.meshgrid(theta, phi)

# Radiation pattern formula for half-wave dipole
R = np.abs(np.sin(theta))

# Convert to Cartesian coordinates
X = R * np.sin(theta) * np.cos(phi)
Y = R * np.sin(theta) * np.sin(phi)
Z = R * np.cos(theta)

# Plotting 3D pattern
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.plot_surface(X, Y, Z)

ax.set_title("3D Radiation Pattern of Dipole Antenna")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

```python
#lab-02
# Isotropic Antenna- 3D 
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 400)
R = np.ones_like(theta) #All values are constant → same radius → circle

ax = plt.subplot(111, projection="polar")
ax.plot(theta, R)
ax.set_title("Isotropic Antenna (2D)")
plt.show()
```

```python
# Isotropic-3D
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, np.pi, 200)
phi = np.linspace(0, 2*np.pi, 200)
theta, phi = np.meshgrid(theta, phi)

R = np.abs(np.cos(theta))

X = R * np.sin(theta) * np.cos(phi)
Y = R * np.sin(theta) * np.sin(phi)
Z = R * np.cos(theta)

ax = plt.figure().add_subplot(projection='3d')
ax.plot_surface(X, Y, Z, cmap="plasma")
ax.set_title("Patch Antenna (3D)")
plt.show()
```

```python
# lab-03
# Horn Antenna-3D (Single directional lobe)
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(-np.pi/2, np.pi/2, 400)
n = 10 # Higher n values create a narrower beam, while lower n values create a wider beam.
R = np.abs(np.cos(theta))**n

ax = plt.subplot(111, projection="polar")
ax.plot(theta, R)
ax.set_title("Horn Antenna (2D)")
plt.show()
```

```python
# Horn Antenna-3D
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, np.pi, 250)
phi = np.linspace(0, 2*np.pi, 250)
theta, phi = np.meshgrid(theta, phi)

n = 6
R = np.abs(np.cos(theta))**n

X = R * np.sin(theta) * np.cos(phi)
Y = R * np.sin(theta) * np.sin(phi)
Z = R * np.cos(theta)

ax = plt.figure().add_subplot(projection='3d')
ax.plot_surface(X, Y, Z, cmap="inferno")
ax.set_title("Horn Antenna (3D)")
plt.show()
```

```python
#lab-04
# Linear Antenna-2D (A set of antenna elements placed in a straight line to improve directivity)
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, np.pi, 800)
N = 10
d = 0.5
k = 2*np.pi

num = np.sin(N * k * d * np.cos(theta) / 2)
den = N * np.sin(k * d * np.cos(theta) / 2)
AF = np.abs(num / den)

ax = plt.subplot(111, projection="polar")
ax.plot(theta, AF)
ax.set_title("Linear Antenna Array (2D)\n")
plt.show()
```

```python
#lab-5: Broadside array antenna 3-element(N=3)
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lam = 1
d = lam / 2
k = 2 * np.pi / lam
N = 3   # number of elements (changed)

# Angle range
theta = np.linspace(0, np.pi, 1000)

# Array Factor (General Form)
AF = np.zeros_like(theta, dtype=complex)

for n in range(N):
    AF += np.exp(1j * n * k * d * np.cos(theta))

# Normalize
AF_mag = np.abs(AF)
AF_mag = AF_mag / np.max(AF_mag)

# Plot
plt.figure(figsize=(6,6))
ax = plt.subplot(111, projection='polar')
ax.plot(theta, AF_mag)
ax.set_title("2D Radiation Pattern (3-Element Broadside Array)\n")
plt.show()
```

```python
#lab-6
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lam = 1.0                 # wavelength (normalized)
d = lam / 2               # element spacing
k = 2 * np.pi / lam       # wave number

# Angle range
theta = np.linspace(0, 2*np.pi, 1000)

# Array Factor (equal amplitude & phase)
AF = 2 * np.cos((k * d / 2) * np.cos(theta))

# Normalized magnitude
AF_norm = np.abs(AF) / np.max(np.abs(AF))

# Plot (Polar)
plt.figure(figsize=(6,6))
plt.polar(theta, AF_norm)
plt.title("Far-Field Pattern\nTwo Isotropic Point Sources\nEqual Amplitude & Phase (d = λ/2)\n\n")
plt.show()
```

```python
#lab-07
# opposite far field pattern
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lam = 1.0                 # wavelength (normalized)
d = lam / 2               # spacing
k = 2 * np.pi / lam       # wave number

# Angle range
theta = np.linspace(0, 2*np.pi, 1000)

# Array Factor (equal amplitude, opposite phase)
AF = 2 * np.sin((k * d / 2) * np.cos(theta))

# Normalized magnitude
AF_norm = np.abs(AF) / np.max(np.abs(AF))

# Polar plot
plt.figure(figsize=(6,6))
plt.polar(theta, AF_norm)
plt.title("Far-Field Pattern\nTwo Isotropic Point Sources\nEqual Amplitude & Opposite Phase (d = λ/2)\n\n")
plt.show()
```

```python
#lab-8
# unequal amplitude any opposite phase
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lam = 1.0                 # wavelength (normalized)
d = lam / 2               # spacing
k = 2 * np.pi / lam       # wave number

A1 = 1.0                  # Amplitude of source 1
A2 = 0.6                  # Amplitude of source 2 (UNEQUAL)

# Angle range
theta = np.linspace(0, 2*np.pi, 1000)

# Phase term
psi = (k * d / 2) * np.cos(theta)

# Array Factor (unequal amplitude, opposite phase)
AF = A1 * np.exp(1j * psi) - A2 * np.exp(-1j * psi)

# Normalized magnitude
AF_norm = np.abs(AF) / np.max(np.abs(AF))

# Polar plot
plt.figure(figsize=(6,6))
plt.polar(theta, AF_norm)
plt.title("Far-Field Pattern\nTwo Isotropic Point Sources\nUnequal Amplitude & Opposite Phase\n\n")
plt.show()
```
